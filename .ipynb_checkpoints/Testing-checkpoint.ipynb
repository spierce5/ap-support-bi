```python
def apiCall(query):
    with open('sf_credentials.yml', 'r') as file:
        cred = yaml.safe_load(file)
    INSTANCE = cred['salesforce_api_login']['instance']
    USERNAME = cred['salesforce_api_login']['username']
    PASSWORD = cred['salesforce_api_login']['password']
    TOKEN = cred['salesforce_api_login']['token']
    sf = Salesforce(instance_url=INSTANCE, username=USERNAME, password=PASSWORD, security_token=TOKEN)
    result = sf.query_all(query)
    
    return result
```

```python


QUERY = '''
SELECT 
    Owner.Name, 
    QS_Severity__c,
    CaseNumber, 
    Account.Name, 
    Type,
    Subject,
    CreatedDate,
    Last_Action_Date__c,
    Contact.Name,  
    Status, 
    Sub_Status__c, 
    LastViewedDate 
FROM Case 
WHERE
    Product_Line_Name__c = 'AssetPoint'
ORDER BY LastViewedDate DESC 
LIMIT 40
'''

result = apiCall(QUERY)

def getName(val):
        if val is None:
            return None
        else:
            return val['Name']

df = pd.DataFrame.from_dict(result['records'])
df['Owner'] = df['Owner'].apply(lambda o: o['Name'])
df['Account'] = df['Account'].apply(lambda o: getName(o))
df['Contact'] = df['Contact'].apply(lambda o: getName(o))
df.drop(columns=['attributes', 'LastViewedDate'], inplace=True)
df.fillna(value="None", inplace=True)
df['Subject'] = df['Subject'].apply(lambda s: s[0:150])
data = list(df.itertuples(index=False, name=None))

df.rename(
    columns={
                'Owner': 'Case Owner',
                'CaseNumber': 'Case Number',
                'Account': 'Account Name',
                'QS_Severity__c': 'Severity',
                'CreatedDate': 'Date/Time Opened',
                'Last_Action_Date__c': 'Last Action Date'
            },
    inplace=True)
```
